```python
import numpy as np
import cv2
import imutils
import matplotlib.pyplot as plt
```

```python
test_img = cv2.imread("../data/training_images/pituitary_tumor/p (33).jpg")
```

```python
test_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB)
```

```python
plt.axis("off")
plt.imshow(test_img)
plt.show()
```

```python
rows, cols, dim = test_img.shape
```

```python
horizontal_reflection = np.float32([[1,0,0],
                                   [0,-1,rows],
                                   [0, 0,1]])
```

```python
hreflection_test_img = cv2.warpPerspective(test_img,
                                           horizontal_reflection,
                                           (int(cols), int(rows)))
```

```python
plt.axis("off")
plt.imshow(hreflection_test_img)
plt.show()
```

```python
vertical_reflection = np.float32([[-1,0,cols],
                                   [0,1,0],
                                   [0, 0,1]])
```

```python
vreflection_test_img = cv2.warpPerspective(test_img,
                                           vertical_reflection,
                                           (int(cols), int(rows)))
```

```python
plt.axis("off")
plt.imshow(vreflection_test_img)
plt.show()
```

```python
rotated_test_image = imutils.rotate(test_img,
                                    angle = 90)
```

```python
plt.axis("off")
plt.imshow(rotated_test_image)
plt.show()
```
